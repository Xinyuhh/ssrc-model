## Data Cleaning

# I. Importing data from data2go NYC

```python
df = pd.read_csv("data2gogo.csv")
df
# 5 - Sector 
# 8 - Variable name
# 15 - Data 
```

# II. Data Cleaning
We only need rows:
5 - Sector data (Contains which sector is this data currently on)
8 - Variable name. What current indicator are we on?
15 - Data itself that matches with column 8

# Steps
1. Extract necessary rows (5, 8, 15)
2. Seperate to Chop off unnecessary rows within 5, 8, 15
3. pivot 
4. join the 3 tables


# Step 1

```python
# 5 - Sector 
# 8 - Variable name
# 15 - Data 
df_copy = df

# Rows I need
final_copy = df.iloc[[5, 8, 15]]
final_copy

# Drop columns 1 and 2 from the final_copy DataFrame
# Chopping off unnecessary columns
final_copy = final_copy.drop(columns=[final_copy.columns[1], final_copy.columns[2]])

# Optionally reset the index if you want it to start from 0
final_copy.reset_index(drop=True, inplace=True)
final_copy
```

# Step 2 Separate each row into its own dataframe

```python
# Create separate DataFrames for each row
sector_df = final_copy.iloc[0:1].copy()  # Row 1 - Sector
label_df = final_copy.iloc[1:2].copy()   # Row 2 - Variable name
data_df = final_copy.iloc[2:3].copy()    # Row 3 - Data

# Optionally reset the index for each DataFrame if you want it to start from 0
sector_df.reset_index(drop=True, inplace=True)
label_df.reset_index(drop=True, inplace=True)
data_df.reset_index(drop=True, inplace=True)

# Display the individual DataFrames (optional, for Jupyter environments)
from IPython.display import display
display(sector_df)
display(label_df)
display(data_df)
```

```python
# Remove first column because there were more unnecessary columns
sector_df = sector_df.iloc[:, 1:]  # Removes the first column from sector_df
label_df = label_df.iloc[:, 1:]    # Removes the first column from label_df
data_df = data_df.iloc[:, 1:]      # Removes the first column from data_df
```

```python
display(label_df)
display(data_df)
display(sector_df)
```

# Step 3 - Pivot each table
Before we pivot and join we have to make sure they are of same size

```python
sector_df.shape
```

```python
label_df.shape
```

```python
data_df.shape
```

```python
# Step 3 & 4 : Pivot and Join
```

```python
# Transpose the data frames
data_df = data_df.T
sector_df = sector_df.T
label_df = label_df.T

# Rename the columns
data_df.columns = ["data"]
sector_df.columns = ["sector"]
label_df.columns = ["label"]

# Concatenate the data frames
final = pd.concat([label_df, data_df, sector_df], axis=1)

# Reset the index to have integers as the row indices (0, 1, 2, ..., 345)
final.reset_index(inplace=True, drop=True)

# Optionally, you can reset the column names to have sequential integers as well
final.reset_index(drop=True, inplace=True)
final.columns = ['Variable/Indicator', 'Data', 'Sector']


# Print the final data frame
print(final)
```

# IMPORTING BACK TO Excel and CSV

```python
import openpyxl

final.to_excel('final_data.xlsx', index=False, header=['Variable/Indicator', 'Data', 'Sector'])
final.to_csv('final_data.csv', index=False, header=['Variable/Indicator', 'Data', 'Sector'])

```
